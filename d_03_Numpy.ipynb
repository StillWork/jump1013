# numpy
- 데이터프레임은 테이블 구조의 데이터를 편리하게 조작하기 위해서 사용한다
- 숫자만으로 구성된 데이터를 대상으로 "연산"을 하기 위해서는 배열 (array) 형태여야  한다
- 이를 위해 넘파이 라이브러리를 사용한다
 - 다 차원 배열 ndarray(n-dimensional array)를 다룬다

## 리스트/데이터프레임과 배열의 차이
- 리스트는 데이터를 "목록"으로 만들거나 조작하는 용도 (추가, 삭제, 변형 등)
- 데이터프레임은 2차원 테이블 구조의 데이터를 만들고, 조작하는 용도
- 배열은 연산을 위해, 숫자만으로 구성된, 수학의 매트릭스와 같은 용도

## 데이터프레임 그래프
- 데이터프레임에서 바로 그래프를 그릴 수 있다

```python
# 데이터프레임 만들기
x1 = {'city': ['서울', '부산', '대구', '대전', '광주'],
        'population': [990, 350, 250, 154, 150],
        'temp': [13, 16, 14, 13, 15]}
x2 = {'city': ['인천', '울산'],
        'population': [290, 120],
        'temp': [12.7, 15.5]}
df1 = pd.DataFrame(x1)
df2 = pd.DataFrame(x2)
df = pd.concat((df1, df2), ignore_index=True)
df
```

```python
# 인구를 백만명 단위로 표시
df.population = df.population/100
df
```

### plot()

```python
# 데이터프레임 내의 수치형 컬럼을 모두 그린다
ax = df.plot()
ax.set_xticks(df.index)
ax.set_xticklabels(df.city)
plt.show()
```

```python
# 특정 컬럼만 보기
ax = df["temp"].plot()
ax.set_xticks(df.index)
ax.set_xticklabels(df.city)
plt.show()
```

### boxplot()

```python
# 데이터프레임에 있는 수치형 컬럼을 모두 그린다
df.boxplot()
```

## plot()에서 kind 를 사용하는 방법
-plot() kind 옵션
 - line, bar, hist, box, kde, area, oie, scatter, hexbin

```python
ax = df.plot(kind='bar')
ax.set_xticks(df.index)
ax.set_xticklabels(df.city)
plt.show()
```

```python
ax = df.plot(kind='barh', stacked=True)
ax.set_yticks(df.index)
ax.set_yticklabels(df.city)
plt.show()
```

```python
ax = df.plot(kind='area', stacked=True, alpha=0.3)
ax.set_xticks(df.index)
ax.set_xticklabels(df.city)
plt.show()
```

## 임의의 x-y 관계 함수 그리기
- 일정 범위 값을 등간격으로 나누는 함수 np.linspace()를 이용
- 공간상 점들을 연결한다

```python
# 등간격으로 나눈 값을 얻는다
xs=np.linspace(0,10,21) 
xs
```

```python
# 임의의 x-y 관계식을 그래프로 그리기
plt.plot(xs, 0.2*xs-1, c='r') # y = 0.2 * x - 1
plt.plot(xs, np.sin(xs), c='k') # y = sin(x)
```
